# Analysing for a particular Year-2003

```python
print(cp.Crop_Year.unique())      #for how many year the data is available and count
print(len(cp.Crop_Year.unique()))
```

```python
cp_2003=cp.loc[cp.Crop_Year==2003]         #As all data is form tamilnadu so the state name column is not needed
cp_2003=cp_2003.drop(['State_Name'],axis=1)
cp_2003
```

```python
#production for each district in 2003

plt.figure(figsize=(20, 10))
sns.barplot(ci=False,
    data=cp_2003,
    x='District_Name',
    y='Production'
)
plt.title("Total production for each district in 2003")
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(ci=False,
    data=cp_2003,
    x='District_Name',
    y='Area'
)
plt.title("Total production for each district in 2003")
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(20, 20))
sns.barplot(ci=False,
    data=cp_2003,
    x='Area',
    y='Crop'
)
plt.title("Total Area in hectare used in 2003 for production");
```

```python
plt.figure(figsize=(20, 20))
sns.barplot(ci=False,
    data=cp_2003,
    x='Production',
    y='Crop'
)
plt.title("production of each crop in 2003");
```

```python
#filtering for Kharif and Rabi season only

cp_03_sea=cp_2003.loc[cp_2003.Season!='Whole Year']
cp_03_sea
```

```python
#comparing Kahrif and Rabi season production in 2003

plt.figure(figsize=(10, 5))
sns.lineplot(ci=False,
    data=cp_03_sea,
    x='Season',
    y='Production'
)
plt.title("Production for each season in 2003");
```

```python
#Area used for those two season

plt.figure(figsize=(10, 5))
sns.barplot(ci=False,
    data=cp_03_sea,
    x='Season',
    y='Area'
)
plt.title("Production for each season");
```

# Gathering for each season

```python
cp.Season.unique()
```

```python
b=cp.groupby('Season')
temp1=b.size()
temp1
```

```python
b.Production.agg(['min','max','mean'])
```

```python
temp2=b.Crop.value_counts()
# type(temp2.Kharif.values)
d1=list(temp2.Kharif.index)
d1v=list(temp2.Kharif.values)
```

```python
#To see which crop is cultivated in kharif season alone

plt.figure(figsize=(20, 20))
plt.pie(d1v,labels=d1)#autopct='%.0f%%'
plt.title('Prodution in preMonsoon (Kharif)')
plt.legend()
```

```python
#whole year season production for all crop types


d2=list(temp2['Whole Year'].index)
d2v=list(temp2['Whole Year'].values)

plt.figure(figsize=(15,20))
plt.plot(d2v,d2)
plt.title('Whole Year production ')
plt.show()
```

# Study of rice prodution during 1997 to 2008

```python
cp_rice=cp.loc[cp.Crop=='Rice']
cp_rice
```

```python
cp_rice['Season']=='Rabi' #no Rice production in winter season
```

```python
#Rice production in each season

plt.figure(figsize=(10, 10))
sns.histplot(
    data=cp_rice,
    x='Season',
    y='Production'
)
plt.title("Rice production in each season");
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(
    data=cp_rice,
    x='Crop_Year',
    y='Production'
)

plt.title("Rice production from 1998-2012 in Tamil nadu");
```

```python
#District wise Rice prodution
plt.figure(figsize=(12, 15))
sns.barplot(
    data=cp_rice,
    y='District_Name',
    x='Production'
)
plt.title("Rice production in each district of Tamil nadu");
```

```python
#Amount of area used and production yield
plt.figure(figsize=(20,10))
sns.lineplot(
    data=cp_rice,
    y='Area',
    x='Production'
)
plt.title("Area use in hectare to production");
```
