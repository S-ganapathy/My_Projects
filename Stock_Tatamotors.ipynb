```python
import yfinance as yf
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Tata_df = yf.download('TATAMOTORS.NS')
```

```python
Tata_df.head()
```

```python
Tata_df.info()
```

```python
# Sampling data by duration
from datetime import date
y=date.today()
# from dateutil.relativedelta import relativedelta
# x=y-relativedelta(years=5)
t=y.year
x=date(t-5,1,1)
print(y,x)
```

```python
Tata_df = Tata_df.loc[x:y]
Tata_df.head()
```

```python
Tata_df.tail()
```

```python
# correlation between the fields
df=Tata_df.corr()
```

```python
sns.heatmap(data=df,annot=True)
```

Removing date from index to a field

```python
Tata_df=Tata_df.reset_index()
```

```python
#Tata_df['Date']=pd.to_datetime(Tata_df.Date)
Tata_df.info()
```

```python
#Adding Month  year quater to the dataframe
#date
Tata_df['Day']=Tata_df.Date.dt.day
#Month
Tata_df['Month']=Tata_df.Date.dt.month
#year
Tata_df['Year']=Tata_df.Date.dt.year
#quater
Tata_df['quad']=Tata_df.Date.dt.quarter

```

```python

# plt.figure(figsize=(10, 5))
sns.set()
figure, axes=plt.subplots(1,2,sharex=True,figsize=(13,5))
figure.suptitle("Market summary")
axes[0].set_title('Open')
axes[1].set_title('Close')

sns.barplot(ci=False,
    data=Tata_df,
    x='Year',
    y='Open',
    ax=axes[0]
)

sns.barplot(ci=False,
    data=Tata_df,
    x='Year',
    y='Close',
    ax=axes[1]
)

```

```python
sns.set()
figure, axes=plt.subplots(1,2,sharex=True,figsize=(13,5))
figure.suptitle("Market summary")
axes[0].set_title('High')
axes[1].set_title('Low')

sns.barplot(ci=False,
    data=Tata_df,
    x='Year',
    y='High',
    ax=axes[0]
)

sns.barplot(ci=False,
    data=Tata_df,
    x='Year',
    y='Low',
    ax=axes[1]
)
```

```python
sns.set()
figure, axes=plt.subplots(2,2,sharex=True,figsize=(13,8))
figure.suptitle("Compare by Days")
axes[0,0].set_title('High')
axes[0,1].set_title('Low')
axes[1,0].set_title('Open')
axes[1,1].set_title('Close')


sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Day',
    y='High',
    ax=axes[0,0]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Day',
    y='Low',
    ax=axes[0,1]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Day',
    y='Open',
    ax=axes[1,0]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Day',
    y='Close',
    ax=axes[1,1]
)
```

```python
sns.set()
figure, axes=plt.subplots(2,2,sharex=True,figsize=(13,8))
figure.suptitle("Compare by Months")
axes[0,0].set_title('High')
axes[0,1].set_title('Low')
axes[1,0].set_title('Open')
axes[1,1].set_title('Close')


sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Month',
    y='High',
    ax=axes[0,0]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Month',
    y='Low',
    ax=axes[0,1]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Month',
    y='Open',
    ax=axes[1,0]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Month',
    y='Close',
    ax=axes[1,1]
)
```

```python
sns.set()
figure, axes=plt.subplots(2,2,sharex=True,figsize=(13,8))
figure.suptitle("Compare by Quater")
axes[0,0].set_title('High')
axes[0,1].set_title('Low')
axes[1,0].set_title('Open')
axes[1,1].set_title('Close')


sns.lineplot(
    ci=False,
    data=Tata_df,
    x='quad',
    y='High',
    ax=axes[0,0]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='quad',
    y='Low',
    ax=axes[0,1]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='quad',
    y='Open',
    ax=axes[1,0]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='quad',
    y='Close',
    ax=axes[1,1]
)
```

```python
sns.set()
figure, axes=plt.subplots(2,2,sharex=True,figsize=(13,8))
figure.suptitle("Compare by Year")
axes[0,0].set_title('High')
axes[0,1].set_title('Low')
axes[1,0].set_title('Open')
axes[1,1].set_title('Close')


sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Year',
    y='High',
    ax=axes[0,0]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Year',
    y='Low',
    ax=axes[0,1]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Year',
    y='Open',
    ax=axes[1,0]
)

sns.lineplot(
    ci=False,
    data=Tata_df,
    x='Year',
    y='Close',
    ax=axes[1,1]
)
```
